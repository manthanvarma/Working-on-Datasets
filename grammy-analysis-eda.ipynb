```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_main = pd.read_csv('/kaggle/input/grammy-award-nominees-and-winners-1958-2024/Grammy Award Nominees and Winners 1958-2024.csv')
df_supp = pd.read_csv('/kaggle/input/grammy-award-nominees-and-winners-1958-2024/Supplementary Table Producer of the Year 2019-2024.csv')
```

```python
def explore_dataset(df, name):
    print(f"\nðŸ“˜ Overview of {name}")
    print("="*60)
    print(f"ðŸ”¹ Shape: {df.shape[0]} rows Ã— {df.shape[1]} columns")
    print(f"ðŸ”¹ Columns: {df.columns.tolist()}")
    print("\nðŸ”¹ First 5 rows:")
    display(df.head())
    print("\nðŸ”¹ Missing values:")
    print(df.isnull().sum())
    print("\nðŸ”¹ Data types:")
    print(df.dtypes)
    print("\nðŸ”¹ Statistical Summary (Numerical Columns):")
    display(df.describe())
```

```python
explore_dataset(df_main, "Main Grammy Dataset (1958â€“2024)")
explore_dataset(df_supp, "Producer of the Year Dataset (2019â€“2024)")
```

```python
df_main['Year'] = pd.to_numeric(df_main['Year'], errors='coerce')
df_supp['Year'] = pd.to_numeric(df_supp['Year'], errors='coerce')
```

```python
print("ðŸ”¸ Unique Award Categories in Main Dataset:")
unique_categories_main = df_main['Award Name'].dropna().unique()
print(f"Total: {len(unique_categories_main)} categories\n")
print(unique_categories_main)
```

```python
print("\nðŸ”¸ Top 15 Categories by Nomination Count:")
top_categories = df_main['Award Name'].value_counts().head(15)
display(top_categories)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y=top_categories.index, x=top_categories.values, palette='viridis')
plt.title("Top 15 Grammy Categories by Number of Nominations")
plt.xlabel("Number of Nominations")
plt.ylabel("Award Name")
plt.tight_layout()
plt.show()
```

```python
categories_per_year = df_main.groupby('Year')['Award Name'].nunique()

plt.figure(figsize=(12, 5))
sns.lineplot(x=categories_per_year.index, y=categories_per_year.values)
plt.title("Number of Unique Grammy Categories Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Categories")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("\nðŸ”¸ Categories in Supplementary Producer Dataset:")
unique_categories_supp = df_supp['Award Name'].dropna().unique()
print(f"Categories: {unique_categories_supp}")
```

```python
df_winners = df_main[df_main['Winner'] == True]

top_winners = df_winners['Nominee'].value_counts().head(15)
print("ðŸ”¸ Top 15 Winning Artists of All Time:")
display(top_winners)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y=top_winners.index, x=top_winners.values, palette='mako')
plt.title("Top 15 Grammy Winners by Win Count")
plt.xlabel("Number of Wins")
plt.ylabel("Artist/Nominee")
plt.tight_layout()
plt.show()
```

```python
nominations_per_artist = df_main['Nominee'].value_counts()
wins_per_artist = df_winners['Nominee'].value_counts()

win_ratio = (wins_per_artist / nominations_per_artist).dropna()
win_ratio = win_ratio[nominations_per_artist >= 5].sort_values(ascending=False).head(15)

print("\nðŸ”¸ Top 15 Artists by Win-to-Nomination Ratio (Min 5 Nominations):")
display(win_ratio)

top_by_category = df_winners.groupby('Award Name')['Nominee'].agg(lambda x: x.value_counts().idxmax())
print("\nðŸ”¸ Most Awarded Artist Per Category (1 winner per category):")
display(top_by_category)
```

```python
awards_per_year = df_main[df_main['Winner'] == True].groupby('Year').size()

nominations_per_year = df_main.groupby('Year').size()

yearly_awards_df = pd.DataFrame({
    'Total Nominations': nominations_per_year,
    'Total Awards (Winners)': awards_per_year
}).fillna(0)

plt.figure(figsize=(14, 6))
sns.lineplot(data=yearly_awards_df, linewidth=2.2)
plt.title("Grammy Nominations and Awards Given Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
max_awards_year = awards_per_year.idxmax()
min_awards_year = awards_per_year.idxmin()

print(f"ðŸ† Year with Most Awards Given: {max_awards_year} ({int(awards_per_year.max())} awards)")
print(f"ðŸ“‰ Year with Least Awards Given: {min_awards_year} ({int(awards_per_year.min())} awards)")
```

```python
df_supp.columns = [col.strip() for col in df_supp.columns]
```

```python
print("\nðŸ”¸ Unique Award Names (Categories) in Producer Dataset:")
print(df_supp['Award Name'].dropna().unique())
```

```python
top_producer_noms = df_supp['Nominee'].value_counts().head(10)
print("\nðŸ”¸ Top 10 Most Nominated Producers (2019â€“2024):")
display(top_producer_noms)
```

```python
producer_winners = df_supp[df_supp['Winner'] == True]
top_producer_wins = producer_winners['Nominee'].value_counts().head(10)
print("\nðŸ”¸ Top 10 Producers by Wins (2019â€“2024):")
display(top_producer_wins)
```

```python
noms = df_supp['Nominee'].value_counts()
wins = producer_winners['Nominee'].value_counts()
producer_win_ratio = (wins / noms).dropna()
producer_win_ratio = producer_win_ratio[noms >= 2].sort_values(ascending=False).head(10)
```

```python
print("\nðŸ”¸ Top 10 Producers by Win-to-Nomination Ratio (Min 2 Nominations):")
display(producer_win_ratio)
```

```python
print("\nðŸ”¸ Producer of the Year Winners by Year and Award Name:")
winners_by_year_cat = producer_winners.groupby(['Year', 'Award Name'])['Nominee'].first()
display(winners_by_year_cat)
```
