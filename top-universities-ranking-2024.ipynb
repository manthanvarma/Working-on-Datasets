```python
# filling missing values with mean

df['N&S'] = df['N&S'].fillna(df['N&S'].mean())
```

```python
# excluding the rows with range values

numeric_df = df.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# top 10 universities by each indicator

#hici

top_10_hici = df[['University_Name', 'Hici']].sort_values(by='Hici', ascending=False).head(10)
top_10_hici

plt.figure(figsize=(10,6))
sns.barplot(x='Hici', y='University_Name', data=top_10_hici, palette='viridis')
plt.title("Top 10 universities by Hici")
plt.plot()
```

```python
#N&S

top_10_nands = df[['University_Name', 'N&S']].sort_values(by='N&S', ascending=False).head(10)
top_10_nands

plt.figure(figsize=(10,6))
sns.barplot(x='N&S', y='University_Name', data=top_10_nands, palette='viridis')
plt.title("Top 10 universities by N&S")
plt.plot()
```

```python
#PUB

top_10_pub = df[['University_Name', 'PUB']].sort_values(by='PUB', ascending=False).head(10)
top_10_pub

plt.figure(figsize=(10,6))
sns.barplot(x='PUB', y='University_Name', data=top_10_pub, palette='viridis')
plt.title("Top 10 universities by PUB")
plt.plot()
```

```python
#PCP

top_10_pcp = df[['University_Name', 'PCP']].sort_values(by='PCP', ascending=False).head(10)
top_10_pcp

plt.figure(figsize=(10,6))
sns.barplot(x='PCP', y='University_Name', data=top_10_pcp, palette='viridis')
plt.title("Top 10 universities by PCP")
plt.plot()
```

```python
# scatter plot for hici and n&s

plt.figure(figsize=(10,6))
sns.scatterplot(x="Hici", y="N&S", data=df)
plt.title("Hici vs N&S")
plt.xlabel('Hici (Highly Cited Researchers)')
plt.ylabel('N&S (Nature and Science Publications)')
plt.plot()
```

```python
# Alumni Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Alumni'], bins=20, kde=True)
plt.title('Distribution of Alumni Scores')
plt.show()
```

```python
# Award Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Award'], bins=20, kde=True)
plt.title('Distribution of Award Scores')
plt.show()
```

```python
# HiCi Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Hici'], bins=20, kde=True)
plt.title('Distribution of HiCi Scores')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Standardizing the data
features = ['Alumni', 'Award', 'Hici', 'N&S', 'PUB', 'PCP']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
```

```python
# Applying KMeans clustering
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Visualizing clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hici', y='PCP', hue='Cluster', data=df, palette='Set2')
plt.title('Clusters of Universities based on HiCi and PCP')
plt.show()
```
