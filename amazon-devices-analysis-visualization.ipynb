```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/phone-search-dataset/phone search.csv')
df.head()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['product_price', 'currency'], inplace=True)
```

```python
df['product_price'] = df['product_price'].replace({'\$': '', ',': ''}, regex=True)
df['product_price'] = pd.to_numeric(df['product_price'], errors='coerce')
df['product_original_price'] = df['product_original_price'].replace({'\$': '', ',': ''}, regex=True)
df['product_original_price'] = pd.to_numeric(df['product_original_price'], errors='coerce')
```

```python
df['product_price'].fillna(df['product_price'].mean(), inplace=True)
```

```python
price_stats = df[['product_price', 'product_original_price']].describe()
print(price_stats)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['product_price'], bins=30, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Product Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['product_price'])
plt.title('Box Plot of Product Prices')
plt.xlabel('Product Price')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='product_price', y='product_star_rating')
plt.title('Price vs. Star Rating')
plt.xlabel('Product Price')
plt.ylabel('Star Rating')
plt.xlim(0, df['product_price'].max() * 1.1)  # Adjust x-axis limit for better visibility
plt.show()
```

```python
df['discount'] = ((df['product_original_price'] - df['product_price']) / df['product_original_price']) * 100

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='discount', y='sales_volume')
plt.title('Discount vs. Sales Volume')
plt.xlabel('Discount Percentage')
plt.ylabel('Sales Volume')
plt.show()
```

```python
def convert_sales_volume(sales):
    if pd.isna(sales) or sales == 'More Buying Choices' or sales == 'List: ' or sales == 'Typical: ':
        return 0  
    if 'K+' in sales:
        return int(sales.split('K+')[0]) * 1000
    elif '+' in sales:
        return int(sales.split('+')[0])  # For cases like '200+'
    return 0

df['sales_volume'] = df['sales_volume'].apply(convert_sales_volume)

print(df[['is_best_seller', 'is_amazon_choice', 'sales_volume']].dtypes)
```

```python
impact_analysis = df.groupby(['is_best_seller', 'is_amazon_choice']).agg(
    average_sales_volume=('sales_volume', 'mean'),
    average_product_price=('product_price', 'mean'),
    average_star_rating=('product_star_rating', 'mean'),
    count=('asin', 'count')
).reset_index()

print(impact_analysis)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=impact_analysis, x='is_best_seller', y='average_sales_volume', hue='is_amazon_choice')
plt.title('Average Sales Volume by Best Seller and Amazon\'s Choice')
plt.xlabel('Is Best Seller')
plt.ylabel('Average Sales Volume')
plt.legend(title='Is Amazon\'s Choice')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=impact_analysis, x='is_best_seller', y='average_product_price', hue='is_amazon_choice')
plt.title('Average Product Price by Best Seller and Amazon\'s Choice')
plt.xlabel('Is Best Seller')
plt.ylabel('Average Product Price')
plt.legend(title='Is Amazon\'s Choice')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=impact_analysis, x='is_best_seller', y='average_star_rating', hue='is_amazon_choice')
plt.title('Average Star Rating by Best Seller and Amazon\'s Choice')
plt.xlabel('Is Best Seller')
plt.ylabel('Average Star Rating')
plt.legend(title='Is Amazon\'s Choice')
plt.show()
```

```python
bestseller_analysis = df.groupby('is_best_seller').agg(
    avg_sales_volume=('sales_volume', 'mean'),
    avg_product_price=('product_price', 'mean'),
    avg_star_rating=('product_star_rating', 'mean')
).reset_index()

amazon_choice_analysis = df.groupby('is_amazon_choice').agg(
    avg_sales_volume=('sales_volume', 'mean'),
    avg_product_price=('product_price', 'mean'),
    avg_star_rating=('product_star_rating', 'mean')
).reset_index()
```

```python
print("Best Seller Analysis:")
print(bestseller_analysis)

print("\nAmazon's Choice Analysis:")
print(amazon_choice_analysis)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.barplot(x='is_best_seller', y='avg_sales_volume', data=bestseller_analysis)
plt.title('Avg Sales Volume by Best Seller Status')
plt.xlabel('Best Seller')
plt.ylabel('Average Sales Volume')

plt.subplot(1, 3, 2)
sns.barplot(x='is_best_seller', y='avg_product_price', data=bestseller_analysis)
plt.title('Avg Product Price by Best Seller Status')
plt.xlabel('Best Seller')
plt.ylabel('Average Product Price')

plt.subplot(1, 3, 3)
sns.barplot(x='is_best_seller', y='avg_star_rating', data=bestseller_analysis)
plt.title('Avg Star Rating by Best Seller Status')
plt.xlabel('Best Seller')
plt.ylabel('Average Star Rating')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.barplot(x='is_amazon_choice', y='avg_sales_volume', data=amazon_choice_analysis)
plt.title("Avg Sales Volume by Amazon's Choice Status")
plt.xlabel("Amazon's Choice")
plt.ylabel("Average Sales Volume")

plt.subplot(1, 3, 2)
sns.barplot(x='is_amazon_choice', y='avg_product_price', data=amazon_choice_analysis)
plt.title("Avg Product Price by Amazon's Choice Status")
plt.xlabel("Amazon's Choice")
plt.ylabel("Average Product Price")

plt.subplot(1, 3, 3)
sns.barplot(x='is_amazon_choice', y='avg_star_rating', data=amazon_choice_analysis)
plt.title("Avg Star Rating by Amazon's Choice Status")
plt.xlabel("Amazon's Choice")
plt.ylabel("Average Star Rating")

plt.tight_layout()
plt.show()
```
