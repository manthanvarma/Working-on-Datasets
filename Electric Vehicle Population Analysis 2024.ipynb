```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.cm as cm
import seaborn as sns
```

```python
cars = pd.read_csv('Datasets/Electric_Vehicle_Data.csv')
```

```python
cars.rename(columns={'County': 'Country'}, inplace=True)
```

```python
colors = ['salmon', 'pink', 'yellow', 'red', 'green', 'orange']
```

```python
# distribution of make

top_makes = cars["Make"].value_counts().head(10)
plt.figure(figsize=(10,4))
top_makes.plot(kind='bar', color = colors)
plt.title('Distribution of Make')
plt.show()
```

```python
# distribution of country

top_makes = cars["Country"].value_counts().head(10)
plt.figure(figsize=(10,4))
top_makes.plot(kind='bar', color = colors)
plt.title('Distribution of Country')
plt.show()
```

```python
# best selling cars at kitsap country

data = cars[cars["Country"] == "Kitsap"]
unique_makes = data["Make"].value_counts().head(10)
unique_makes.plot(kind="bar", color = colors)
plt.title('Best Selling Cars at Kitsap')
plt.figure(figsize=(10,4))
plt.show()
```

```python
# distribution by models
model_data = cars["Model"].value_counts().head(5)
plt.figure(figsize=(10,4))
model_data.plot(kind='bar', color = colors)
plt.title('Distribution by Models')
plt.show()
```

Number of vehicles eligible for Clean Alternative Fuels

```python
data = cars['Clean Alternative Fuel Vehicle (CAFV) Eligibility']
value_counts = data.value_counts()

plt.figure(figsize=(6, 6))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)

plt.title("Number of vehicles eligible for clean alternative fuels")
plt.axis('equal')
plt.show()
```

Electric Range vs Frequency

```python
electric_range_data = cars["Electric Range"]
plt.figure(figsize=(8,6))
sns.histplot(data = electric_range_data, bins=15, kde=True, color="skyblue")
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.title('Power Reserve')
plt.show()
```
