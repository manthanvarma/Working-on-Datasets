```python
import pandas as pd
```

```python
df = pd.read_csv("premier-player-23-24.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
df.columns = df.columns.str.strip()
df = df.drop_duplicates()
df.head()
```

```python
df.to_csv("premier-player-23-24.csv cleaned", index=False)
```

```python
df.describe()
```

```python
# distribution of metrics

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
metrics = ["Gls", "Ast", "xG", "xAG", "npxG"]

plt.figure(figsize=(10,6))
for i, metric in enumerate(metrics, 1):
    plt.subplot(3,2,i)
    sns.histplot(df[metric].dropna(), kde=True, bins=20)
    plt.title(f'Distribution of {metric}')
    plt.xlabel(metric)
    plt.ylabel("frequency")

plt.tight_layout()
plt.show()
```

```python
# correlation metrics

numeric_df = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Player Performance Metrics')
plt.show()
```

```python
# performance by position

position_stats = df.groupby('Pos')[['Gls', 'Ast', 'xG', 'xAG', 'npxG']].mean()

position_stats.plot(kind='bar', figsize=(10,6))
plt.title('Average performance metrics by position')
plt.ylabel('Average values')
plt.xticks(rotation=0)
plt.legend(title='metrics')
plt.show()
```

```python
# age vs performance

plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.scatterplot(x='Age', y='Gls', data=df)
plt.title("age vs goals")
plt.xlabel('Age')
plt.ylabel('Goals')

plt.subplot(1,2,2)
sns.scatterplot(x='Age', y='Ast', data=df)
plt.title("age vs assists")
plt.xlabel('Age')
plt.ylabel('Assists')

plt.tight_layout()
plt.show()
```

```python
# top performers

top_scorers = df[['Player', 'Gls']].sort_values(by='Gls', ascending=False).head(10)
print('Top 10 players by Goals\n')
print(top_scorers, "\n")

top_assists = df[['Player', 'Ast']].sort_values(by='Ast', ascending=False).head(10)
print('Top 10 players by Assists\n')
print(top_assists, "\n")

top_xG = df[['Player', 'xG']].sort_values(by='xG', ascending=False).head(10)
print('Top 10 players by Expected Goals\n')
print(top_xG, "\n")
```

```python
# goals + assists to check which player contributed the most

goalsassists = df[['Player', 'G+A']].sort_values(by='G+A', ascending=False).head(10)
print("Top 10 players by goals and assists")
print(goalsassists) 
```

```python
# goals and assists per 90 minutes

df['Gls_per_90'] = df['Gls'] / df['90s']

df['Ast_per_90'] = df['Ast'] / df['90s']

top_goals_per_90 = df[['Player', 'Gls_per_90']].sort_values(by='Gls_per_90', ascending=False).head(10)

top_assists_per_90 = df[['Player', 'Ast_per_90']].sort_values(by='Ast_per_90', ascending=False).head(10)

print("Top 10 Players by Goals per 90 Minutes:")
print(top_goals_per_90)
print("\nTop 10 Players by Assists per 90 Minutes:")
print(top_assists_per_90)
```

```python
# over or underperformance by comparing expected goals with actual goals and vice versa with assists

df['Gls_minus_xG'] = df['Gls'] - df['xG']
df['Ast_minus_xAG'] = df['Ast'] - df['xAG']

overperform_goals = df[['Player', 'Gls_minus_xG']].sort_values(by='Gls_minus_xG', ascending=False).head(10)

underperform_goals = df[['Player', 'Gls_minus_xG']].sort_values(by='Gls_minus_xG').head(10)

print("Top 10 Players Who Overperformed in Goals:")
print(overperform_goals)
print("\nTop 10 Players Who Underperformed in Goals:")
print(underperform_goals)
```

```python
# top 10 players with progressive carries and passes

top_prgc = df[['Player', 'PrgC']].sort_values(by='PrgC', ascending=False).head(10)

top_prgp = df[['Player', 'PrgP']].sort_values(by='PrgP', ascending=False).head(10)

print("Top 10 Players by Progressive Carries:")
print(top_prgc)
print("\nTop 10 Players by Progressive Passes:")
print(top_prgp)
```

```python
# visualization of above operations

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.scatterplot(x='xG', y='Gls', data=df, alpha=0.7)
plt.title('Actual Goals vs. Expected Goals (xG)')
plt.xlabel('Expected Goals (xG)')
plt.ylabel('Actual Goals (Gls)')
plt.plot([0, df['xG'].max()], [0, df['xG'].max()], 'r--')  
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='xAG', y='Ast', data=df, alpha=0.7)
plt.title('Actual Assists vs. Expected Assists (xAG)')
plt.xlabel('Expected Assists (xAG)')
plt.ylabel('Actual Assists (Ast)')
plt.plot([0, df['xAG'].max()], [0, df['xAG'].max()], 'r--') 
plt.show()
```

```python
df['Gls_minus_xG'] = df['Gls'] - df['xG']
df['Ast_minus_xAG'] = df['Ast'] - df['xAG']

overperform_goals = df[['Player', 'Gls', 'xG', 'Gls_minus_xG']].sort_values(by='Gls_minus_xG', ascending=False).head(10)

underperform_goals = df[['Player', 'Gls', 'xG', 'Gls_minus_xG']].sort_values(by='Gls_minus_xG').head(10)

overperform_assists = df[['Player', 'Ast', 'xAG', 'Ast_minus_xAG']].sort_values(by='Ast_minus_xAG', ascending=False).head(10)

underperform_assists = df[['Player', 'Ast', 'xAG', 'Ast_minus_xAG']].sort_values(by='Ast_minus_xAG').head(10)

print("Top 10 Players Who Overperformed in Goals:")
print(overperform_goals)
print("\nTop 10 Players Who Underperformed in Goals:")
print(underperform_goals)
print("\nTop 10 Players Who Overperformed in Assists:")
print(overperform_assists)
print("\nTop 10 Players Who Underperformed in Assists:")
print(underperform_assists)
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='xG', y='Gls', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Expected Goals (xG) vs. Actual Goals (Gls)')
plt.xlabel('Expected Goals (xG)')
plt.ylabel('Actual Goals (Gls)')
plt.show()

plt.figure(figsize=(12, 6))
sns.regplot(x='xAG', y='Ast', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Expected Assists (xAG) vs. Actual Assists (Ast)')
plt.xlabel('Expected Assists (xAG)')
plt.ylabel('Actual Assists (Ast)')
plt.show()
```

```python
# distribution of yellow and red cards

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.histplot(df['CrdY'].dropna(), kde=True, bins=20, color='orange')
plt.title('Distribution of Yellow Cards')
plt.xlabel('Yellow Cards')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['CrdR'].dropna(), kde=True, bins=20, color='red')
plt.title('Distribution of Red Cards')
plt.xlabel('Red Cards')
plt.ylabel('Frequency')
plt.show()
```

```python
# average performance metrics by number of cards

yellow_card_stats = df.groupby('CrdY')[['Gls', 'Ast']].mean()

yellow_card_stats.plot(kind='bar', figsize=(14, 8))
plt.title('Average Goals and Assists by Yellow Cards')
plt.xlabel('Number of Yellow Cards')
plt.ylabel('Average Values')
plt.xticks(rotation=0)
plt.legend(title='Metrics')
plt.show()

red_card_stats = df.groupby('CrdR')[['Gls', 'Ast']].mean()

red_card_stats.plot(kind='bar', figsize=(14, 8))
plt.title('Average Goals and Assists by Red Cards')
plt.xlabel('Number of Red Cards')
plt.ylabel('Average Values')
plt.xticks(rotation=0)
plt.legend(title='Metrics')
plt.show()
```

```python
# correlation metrics for yellow and red card

correlation_with_yellow = df[['CrdY', 'Gls', 'Ast']].corr()
correlation_with_red = df[['CrdR', 'Gls', 'Ast']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_with_yellow, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation with Yellow Cards')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_with_red, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation with Red Cards')
plt.show()
```
