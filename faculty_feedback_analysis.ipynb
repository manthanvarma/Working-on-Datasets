```python
teacher_stats = df.groupby('Teacher')['Score'].agg(['mean', 'std', 'count']).reset_index()
teacher_stats.sort_values(by='mean', ascending=False)
```

```python
teacher_stats['adjusted_score'] = teacher_stats['mean'] * (teacher_stats['count'] / teacher_stats['count'].max())
teacher_stats.sort_values(by='adjusted_score', ascending=False)
```

```python
def classify(score):
    if score >= 3.5:
        return "Elite"
    elif score >= 3.0:
        return "Good"
    elif score >= 2.5:
        return "Average"
    else:
        return "Needs Improvement"

teacher_stats['Category'] = teacher_stats['mean'].apply(classify)

teacher_stats.sort_values(by='mean', ascending=False)
```

```python
teacher_stats['Category'].value_counts()
```

```python
teacher_stats.sort_values(by='std').head(10)
```

```python
consistent_teachers = teacher_stats[teacher_stats['count'] > 1]
consistent_teachers.sort_values(by='std').head(10)
```

```python
consistent_teachers.sort_values(by='std', ascending=False).head(10)
```

```python
import numpy as np

mean_score = teacher_stats['mean'].mean()
std_score = teacher_stats['mean'].std()

teacher_stats['z_score'] = (teacher_stats['mean'] - mean_score) / std_score
outliers = teacher_stats[np.abs(teacher_stats['z_score']) > 1.5]
outliers
```

```python
from sklearn.cluster import KMeans
X = teacher_stats[['mean', 'std']].fillna(0)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
teacher_stats['Cluster'] = kmeans.fit_predict(X)
teacher_stats
```

```python
teacher_stats.groupby('Cluster')[['mean', 'std']].mean()
```
